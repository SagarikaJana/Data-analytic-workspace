```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [1,12,5,7,8]
```

```python
plt.plot(x,y,label = "Data")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Representation of data")
plt.show()
```

```python
plt.scatter(x,y,label = "Data",color ='r')
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Representation of data")
plt.show()
```

```python
# Histogram
x1 = [76,88,99,55,112,133,143,166,111,120]
plt.hist(x1,bins = 3,rwidth =0.9)
plt.title("Representation of data")
plt.show()
```

```python
# Histogram
x1 = [97,96,99,100,101,102.103,100,99,100]
plt.hist(x1,bins = 3,rwidth =0.9,color='g')
plt.title("Representation of data")
plt.show()
```

```python
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height)
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
 
# Show graphic
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Food', 'Rent', 'Savings', 'Miscellaneous'
sizes = [15, 30, 45, 10]
explode = (0, 0, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Savings')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
