# Working on Pandas with Graphs

```python
# import the library files 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from the directory is crating a dataframe as well
iris = pd.read_csv(r'C:\Users\Analytics Domain\Desktop\Iris1.csv')
# Display the dataframe
iris.head()
```

```python
iris.tail()
```

```python
# counted the number of records
print(iris['iris'].count())
```

```python
print(iris['iris'].value_counts())
```

```python
#create a basic scatter plot for the features 'sepal length' and 'sepal Width'
plt.plot(iris['petal length'],iris['petal width'],marker='.', linestyle='none')
```

```python
# Using Seaborn to create the graph
sns.FacetGrid(iris,hue='iris', size=8).map(plt.scatter,"sepal length","sepal width").add_legend()
```

```python
# using the Join Plot from Seaborn library
sns.jointplot(x="sepal length", y = "sepal width", data = iris, size=5, kind="hex",color="#4CB391")
```

```python
# Create a box plot with seaborn
sns.boxplot(x="iris",y ="sepal width", data = iris)
```

```python
sns.violinplot(x="iris", y="sepal width", data = iris, size =50)
```

```python
sns.FacetGrid(iris, hue="iris", size = 6).map(sns.kdeplot, "petal length").add_legend()
```

```python
sns.violinplot(x="iris", y="petal width", data =iris, size = 150)
               
               
```
